```python
#to find the housing value read the csv file to predict the actual value
import pandas as pd
import numpy as np
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
df.head()
```

```python
#here we taken the x as  input and y as predicted output
x = df.drop('median_house_value',axis=1)
y = df['median_house_value']
```

```python
#import libraries and train,test, split the datas
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
# import KNeighborsTransformer libraries
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
#to train the linear regression  data modles
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Instantiate a KNeighborsRegressor object
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
```

```python
#to fit the trained data modles
model.fit(x_train , y_train)
```

```python
#to correct the prblem by pickle
import pickle
pickle.dump(model,open('model.pkl','wb'))
pickle.dump(model,open('scaler.pkl','wb'))
```

```python
# add  coloumn and row
input_data ={'longitude':-118.0, 'latitude':34.0, 'housing_median_age':30.0, 'total_rooms':6.0, 'total_bedrooms':1.0, 'population':1000.0, 'households':3.0, 'median_income':5.0}
```

```python
#import mean and sk learn
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Make predictions on the test data
y_pred = model.predict(x_test)
```

```python
# Calculate RMSE
from sklearn.metrics import r2_score
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Squared Error: {rmse}")
print(f"R2 score: {r2_score(y_test, y_pred)}")
```
